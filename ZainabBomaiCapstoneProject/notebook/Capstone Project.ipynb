```python
# Importing relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv("Tetuan City power consumption.csv")
```

```python
data
```

```python
data['DateTime'] = pd.to_datetime(data['DateTime'])
```

```python
data['hour'] = data['DateTime'].dt.hour
data['day_of_week'] = data['DateTime'].dt.dayofweek
data['month'] = data['DateTime'].dt.month
```

```python
data
```

```python
data = data.drop('DateTime', axis=1)
```

```python
data.isna().sum()
```

```python
X = data[['Temperature','Humidity','Wind Speed','general diffuse flows','diffuse flows','hour','day_of_week','month']]
```

```python
X
```

```python
Y = data[['Zone 1 Power Consumption']]
```

```python
Y
```

```python
lr1 = linear_model.LinearRegression()
```

```python
lr1
```

```python
lr1.fit(X,Y)
```

```python
lr1.coef_
```

```python
lr1.intercept_
```

```python
lr1.predict([[7.567,75.9,0.090,0.079,0.152,15,5,3]])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
# Step 5: Make predictions
Y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(Y_test, Y_pred)
mae = mean_absolute_error(Y_test, Y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(Y_test, Y_pred)

print("MAPE: ", mean_absolute_percentage_error(Y_test, Y_pred))
print("R² Score:", r2)
print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("Real Mean Squared Error (RMSE): ", rmse)
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor(n_estimators=200, learning_rate=0.05, max_depth=5, random_state=42)

xgb.fit(X_train, Y_train)

Y_pred_xgb = xgb.predict(X_test)
```

```python
mae = mean_absolute_error(Y_test, Y_pred_xgb)
mse = mean_squared_error(Y_test, Y_pred_xgb)
rmse = np.sqrt(mse)

print("XGBoost MAE:", mae)
print("XGBoost MSE:", mse)
print("XGBoost RMSE:", rmse)
print("XGBoost MAPE: ", mean_absolute_percentage_error(Y_test, Y_pred_xgb))
print("R2:", r2_score(Y_test, Y_pred_xgb))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train_scaled, Y_train)
```

```python
Y_pred_knn = knn.predict(X_test_scaled)
```

```python
mse = mean_squared_error(Y_test, Y_pred_knn)
rmse = np.sqrt(mse)

print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mean_absolute_error(Y_test, Y_pred_knn))
print("R2:", r2_score(Y_test, Y_pred_knn))
print("MAPE: ", mean_absolute_percentage_error(Y_test, Y_pred_knn))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators = 100, random_state = 42)
```

```python
rf.fit(X_train, Y_train)
```

```python
Y_pred_rf = rf.predict(X_test)
```

```python
mse = mean_squared_error(Y_test, Y_pred_rf)
rmse = np.sqrt(mse)

print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mean_absolute_error(Y_test, Y_pred_rf))
print("R2:", r2_score(Y_test, Y_pred_rf))
print("MAPE: ", mean_absolute_percentage_error(Y_test, Y_pred_rf))
```
